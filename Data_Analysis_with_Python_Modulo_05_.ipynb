#**Model evaluation**

La evaluacin del modelo nos dice cmo funciona nuestro modelo en el mundo real. En el mdulo anterior, se habl sobre la evaluacin en la muestra. La evaluacin de la muestra nos dice qu tan bien se ajusta nuestro modelo a los datos ya proporcionados para entrenarlo. No nos da una estimacin de qu tan bien el modelo entrenado puede predecir nuevos datos. La solucin es dividir nuestros datos, usar los datos de la muestra o los datos de entrenamiento para entrenar el modelo. El resto de los datos, denominados datos de prueba, se utilizan como datos fuera de la muestra. Estos datos se utilizan luego para aproximar el rendimiento del modelo en el mundo real. Separar los datos en conjuntos de entrenamiento y prueba es una parte importante de la evaluacin del modelo. Usamos los datos de prueba para tener una idea de cmo funcionar nuestro modelo en el mundo real. 

Cuando dividimos un conjunto de datos, generalmente la mayor parte de los datos se usa para entrenamiento y una parte ms pequea se usa para pruebas. Por ejemplo, podemos usar el 70% de los datos para entrenamiento; luego usamos el 30% para las pruebas. Usamos un conjunto de entrenamiento para construir un modelo y descubrir relaciones predictivas. Luego usamos un conjunto de prueba para evaluar el rendimiento del modelo. Cuando hayamos completado la prueba de nuestro modelo, debemos usar todos los datos para entrenar el modelo. 

image.png

Una funcin popular en el paquete de aprendizaje sci-kit para dividir conjuntos de datos es la funcin de "train-test-split". Esta funcin divide aleatoriamente un conjunto de datos en subconjuntos de entrenamiento y prueba. Del fragmento de cdigo de ejemplo, este mtodo se importa de "sklearn.cross_validation". Los parmetros de entrada y_data es la variable objetivo (en el ejemplo de la valoracin del coche, sera el precio), y "x_data", la lista de variables predictoras. En este caso, seran todas las dems variables en el conjunto de datos del automvil las que estamos usando para intentar predecir el precio. La salida es una matriz: "x_train" y "y_train", los subconjuntos para el entrenamiento; "x_test" y "y_test", los subconjuntos para la prueba. En este caso, el porcentaje de "tamao de prueba" de los datos para el conjunto de prueba. Aqu est el 30%. El "random_state" es una semilla aleatoria para la divisin aleatoria de conjuntos de datos. El error de generalizacin es una medida de qu tan bien nuestros datos predicen datos nunca antes vistos. El error que obtenemos usando nuestros datos de prueba es una aproximacin de este error. 

image.png

El uso de una gran cantidad de datos para el entrenamiento nos brinda un medio preciso para determinar cmo funcionar nuestro modelo en el mundo real, pero la precisin del rendimiento ser baja. La primera vez que experimentamos obtenemos una buena estimacin de los datos de entrenamiento. Si experimentamos nuevamente, entrenando el modelo con una combinacin diferente de muestras, tambin obtendremos un buen resultado, pero los resultados sern diferentes en relacin con la primera vez que ejecutamos el experimento. Repitiendo el experimento nuevamente con una combinacin diferente de muestras de entrenamiento y prueba, los resultados son relativamente cercanos al error de feneralizacin, pero distintos entre s. Repitiendo el proceso, obtenemos una buena aproximacin del error de generalizacin, pero la precisin es pobre, es decir, todos los resultados son extremadamente diferentes entre s. Si usamos menos puntos de datos para entrenar el modelo y ms para probar el modelo, la precisin del rendimiento de la generalizacin ser menor, pero el modelo tendr una buena precisin. 

Una de las mtricas de evaluacin fuera de muestra ms comunes es la validacin cruzada. En este mtodo, el conjunto de datos se divide en k-grupos iguales; cada grupo se conoce como un pliegue. Por ejemplo 4 segmentos. Algunos de los segmentos se pueden usar como un conjunto de entrenamiento, que usamos para entrenar el modelo, y las partes restantes se usan como un conjunto de prueba, que usamos para probar el modelo. Por ejemplo, podemos usar tres segmentos para entrenar; luego usaremos un segmento para probar. Esto se repite hasta que cada particin se usa tanto para entrenamiento como para prueba. Al final, usamos los resultados promedio como la estimacin del error fuera de la muestra. La mtrica de evaluacin depende del modelo. Por ejemplo, el R-cuadrado. La forma ms sencilla de aplicar la validacin cruzada es llamar a la funcin cross_val_score(), que realiza mltiples evaluaciones 'fuera de la muestra'. Este mtodo se importa del paquete de seleccin de modelos de sklearn. Luego usamos la funcin cross_val_score().

El primer parmetro de entrada es el tipo de modelo que estamos usando para hacer la validacin cruzada. En este ejemplo, inicializamos un modelo de regresin lineal u objeto lr, que pasamos a la funcin cross_val_score. Los otros parmetros son x_data, los datos de la variable de prediccin, y y_data, los datos de la variable de destino. Podemos gestionar el nmero de particiones con el parmetro cv. Aqu, cv = 3, lo que significa que el conjunto de datos se divide en 3 particiones iguales. La funcin devuelve una matriz de puntuaciones, una para cada particin que se eligi como conjunto de prueba. Podemos promediar el resultado juntos para estimar R-cuadrado fuera de la muestra usando la funcin media en numpy. 

#**Overfitting, Underfitting and Model Selection**

Considere la siguiente funcin: 

y(x) + noise

image.png

asumimos que los puntos de entrenamiento provienen de una funcin polinomial ms algo de ruido. El objetivo de la seleccin del modelo es determinar el orden del polinomio para proporcionar la mejor estimacin de la funcin y(x). Si tratamos de ajustar la funcin con una funcin lineal, la lnea no es lo suficientemente compleja para ajustar los datos. Como resultado, hay muchos errores. Esto se denomina subajuste, cuando el modelo es demasiado simple para ajustarse a los datos. Si aumentamos el orden del polinomio, el modelo se ajusta mejor, pero el modelo an no es lo suficientemente flexible y presenta un ajuste insuficiente. 

Al aumentar a un polinomio de orden 16, el modelo funciona extremadamente bien en el seguimiento de los puntos de entrenamiento, pero funciona mal en la estimacin de la funcin. Esto es especialmente evidente cuando hay pocos datos de entrenamiento; la funcin estimada oscila sin seguir la funcin. Esto se denomina sobreajuste, donde el modelo es demasiado flexible y se ajusta al ruido en lugar de a la funcin. 

Si vemos un grfico del error cuadrtico medio para el conjunto de entrenamiento y prueba para polinomios de diferente orden. El eje horizontal representa el orden del polinomio; el eje vertical es el error cuadrtico medio. El error de entrenamiento disminuye con el orden del polinomio. El error de prueba es un mejor medio para estimar el error de un polinomio. El error disminuye hasta que se determina el mejor orden del polinomio, luego el error comienza a aumentar. Seleccionamos el orden que minimiza el error de prueba, en este caso, fue 8. Cualquier cosa a la izquierda se considerara de ajuste insuficiente. Cualquier cosa a la derecha es demasiado adecuada. Si seleccionamos el mejor orden del polinomio, an tendremos algunos errores, si recuerdas, la expresin original de los puntos de entrenamiento. Vemos un trmino de ruido; este trmino es una de las razones del error. Esto se debe a que el ruido es aleatorio y no podemos predecirlo; esto a veces se denomina error irreducible.

#**Ridge Regression**

La Ridge Regression evita el sobreajuste. El ajuste excesivo tambin es un gran problema cuando tiene varias variables o caractersticas independientes. 

Considere el siguiente polinomio de cuarto orden en naranja. Los puntos azules se generan a partir de esta funcin. Podemos usar un polinomio de dcimo orden para ajustar los datos. La funcin estimada en azul hace un buen trabajo al aproximar la funcin verdadera. En muchos casos, los datos reales tienen valores atpicos. Por ejemplo, este punto que se muestra aqu no parece provenir de la funcin en naranja. Si usamos una funcin polinomial de dcimo orden para ajustar los datos, la funcin estimada en azul es incorrecta y no es una buena estimacin de la funcin real en naranja. Si examinamos la expresin de la funcin estimada, vemos que los coeficientes polinmicos estimados tienen una magnitud muy grande. 

image.png

Esto es especialmente evidente para los polinomios de orden superior. La Ridge Regression controla la magnitud de estos coeficientes polinmicos introduciendo el parmetro alfa. Alfa es un parmetro que seleccionamos antes de ajustar o entrenar el modelo. Cada fila de la siguiente tabla representa un valor creciente de alfa. Veamos cmo diferentes valores de alfa cambian el modelo. Esta tabla representa los coeficientes polinmicos para diferentes valores de alfa. Las columnas corresponden a los diferentes coeficientes polinmicos y las filas corresponden a los diferentes valores de alfa. A medida que aumenta alfa, los parmetros se hacen ms pequeos. Esto es ms evidente para las caractersticas polinmicas de orden superior, pero alfa debe seleccionarse con cuidado. Si alfa es demasiado grande, los coeficientes se aproximarn a cero y se ajustarn por debajo de los datos. Si alfa es cero, el sobreajuste es evidente. Para alfa igual a 0,001, el sobreajuste comienza a disminuir. Para alfa igual a 0,01, la funcin estimada sigue la funcin real. Cuando alfa es igual a 1, vemos los primeros signos de subajuste. La funcin estimada no tiene suficiente flexibilidad. En alfa igual a 10, vemos un ajuste insuficiente extremo; ni siquiera rastrea los dos puntos. Para seleccionar alfa usamos validacin cruzada. 

image.png

Para hacer una prediccin mediante la Ridge Regression, importe Ridge de modelos lineales de sklearn. Cree un objeto Ridge usando el constructor. El parmetro alfa es uno de los argumentos del constructor. Entrenamos el modelo usando el mtodo de ajuste. Para hacer una prediccin, usamos el mtodo de prediccin. Para determinar el parmetro alfa, usamos algunos datos para el entrenamiento. Usamos un segundo conjunto llamado datos de validacin; esto es similar a los datos de prueba, pero se usa para seleccionar parmetros como alfa. Comenzamos con un pequeo valor de alfa, entrenamos el modelo, hacemos una prediccin utilizando los datos de validacin, luego calculamos el R cuadrado y almacenamos los valores. 

Repita el valor para un valor mayor de alfa. Entrenamos el modelo nuevamente, hacemos una prediccin usando los datos de validacin, luego calculamos el R cuadrado y almacenamos los valores de R cuadrado. Repetimos el proceso para un valor alfa diferente, entrenando el modelo y haciendo una prediccin. Seleccionamos el valor de alfa que maximiza la R al cuadrado. Tenga en cuenta que podemos usar otras mtricas para seleccionar el valor de alfa como el error cuadrtico medio. El problema del sobreajuste es an peor si tenemos muchas caractersticas. La siguiente grfica muestra los diferentes valores de R al cuadrado en el eje vertical. El eje horizontal representa diferentes valores para alfa. Usamos varias caractersticas de nuestro conjunto de datos de autos usados y una funcin polinomial de segundo orden. Los datos de entrenamiento estn en rojo y los datos de validacin en azul. Vemos que a medida que aumenta el valor de alfa, aumenta el valor de R al cuadrado y converge en aproximadamente 0,75. En este caso, seleccionamos el valor mximo de alfa porque ejecutar el experimento para valores ms altos de alfa tiene poco impacto. Por el contrario, a medida que aumenta alfa, disminuye el R cuadrado de los datos de prueba. Esto se debe a que el trmino alfa evita el sobreajuste. Esto puede mejorar los resultados en los datos no vistos, pero el modelo tiene un peor rendimiento en los datos de prueba. Consulte el laboratorio sobre cmo generar este grfico.

#**Grid Search**

El Grid Search nos permite escanear a travs de mltiples parmetros con pocas lneas de cdigo. Los parmetros como el trmino alfa discutido, no son parte del proceso de adaptacin o entrenamiento. Estos valores se denominan hiperparmetros. Scikit-learn tiene un medio para iterar automticamente sobre estos hiperparmetros mediante validacin cruzada. Este mtodo se llama Grid Search. 

image.png

El Grid Search toma el modelo o los objetos que le gustara entrenar y diferentes valores de los hiperparmetros. Luego calcula el error cuadrtico medio o R cuadrado para varios valores de hiperparmetros, lo que le permite elegir los mejores valores. Usamos diferentes hiperparmetros para entrenar el modelo. Continuamos el proceso hasta agotar los diferentes valores de parmetros. Cada modelo produce un error. Seleccionamos el hiperparmetro que minimiza el error. Para seleccionar el hiperparmetro, dividimos nuestro conjunto de datos en tres partes, el conjunto de entrenamiento, el conjunto de validacin y el conjunto de prueba. Entrenamos el modelo para diferentes hiperparmetros. Usamos la R cuadrada o el error cuadrtico medio para cada modelo. Seleccionamos el hiperparmetro que minimiza el error cuadrtico medio o maximiza el R cuadrado en el conjunto de validacin. Finalmente probamos el rendimiento de nuestro modelo utilizando los datos de prueba. 

image.png

El valor de un Grid Search es una lista de Python que contiene un diccionario de Python. La clave es el nombre del parmetro libre. El valor del diccionario son los diferentes valores del parmetro libre. Esto se puede ver como una tabla con varios valores de parmetros libres. Tambin tenemos el objeto o modelo. El Grid Search asume el mtodo de puntuacin, en este caso R cuadrado, el nmero de pliegues, el modelo u objeto y los valores de los parmetros libres. Algunas de las salidas incluyen las diferentes puntuaciones para diferentes valores de parmetros libres; en este caso la R al cuadrado junto con los valores de los parmetros libres que tengan mejor puntuacin. Primero, importamos las bibliotecas que necesitamos, incluido Grid Search CV, el diccionario de valores de parmetros.

```python
# Funciones para plotear

def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Proportion of Cars')

    plt.show()
    plt.close()
```

```python
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    
    #training data 
    #testing data 
    # lr:  linear regression object 
    #poly_transform:  polynomial transformation object 
 
    xmax=max([xtrain.values.max(), xtest.values.max()])

    xmin=min([xtrain.values.min(), xtest.values.min()])

    x=np.arange(xmin, xmax, 0.1)


    plt.plot(xtrain, y_train, 'ro', label='Training Data')
    plt.plot(xtest, y_test, 'go', label='Test Data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Predicted Function')
    plt.ylim([-10000, 60000])
    plt.ylabel('Price')
    plt.legend()
```

```python
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```
