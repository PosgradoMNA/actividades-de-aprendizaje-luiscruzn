**Definición de Latitud y Longitud**

**Latitud** Es la distancia en grados, minutos y segundos que hay con respecto al paralelo principal, que es el ecuador (0º). La latitud puede ser norte y sur.

**Longitud**: Es la distancia en grados, minutos y segundos que hay con respecto al meridiano principal, que es el meridiano de Greenwich (0º).La longitud puede ser este y oeste.

```python
from sklearn.cluster import KMeans
from geopy.geocoders.yandex import Location
from geopy.geocoders import Nominatim 
from geopy.distance import geodesic
```

```python
num_clusters = range(1,10)
scores = list()

for k in num_clusters:
    kmeans = KMeans(n_clusters=k)
    estimator = kmeans.fit(latlong[['longitude']])
    scores.append(estimator.score(latlong[['longitude']]))

plt.plot(num_clusters, scores)
plt.xlabel('Número de Clusters')
plt.ylabel('Score')
plt.title('Gŕafica de codo')
plt.xticks(np.arange(min(num_clusters), max(num_clusters)+1, 1.0))
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state=41)
kmeans.fit(latlong[['longitude','latitude']])

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print("=============Labels================")
print(np.unique(labels))
print("=============Centroides============")
print(centroids)
```

```python
centroids_df = pd.DataFrame(centroids) 
centroids_df["latlon"] = list(zip(centroids_df[0], centroids_df[1]))
centroids_df["latlon"] = centroids_df["latlon"].apply(Point)
centroids_df["address"] = ""

geodataframe = gpd.GeoDataFrame(centroids_df, geometry="latlon") 
geodataframe
```

```python
fig, gax = plt.subplots(figsize=(15,10))

# By only plotting rows in which the continent is 'South America' we only plot, well,
# South America.
world.query("name == 'United States of America'").plot(ax = gax, edgecolor='black', color='white') #Aqui traemos el mapa de los unites

# This plot the cities. It's the same syntax, but we are plotting from a different GeoDataFrame.
# I want the cities as pale red dots.
gdf.plot(ax=gax, color='blue', alpha = 0.5) #Aqui grafica los datos originales
geodataframe.plot(ax=gax, color='black', alpha = 1, markersize = 300) #Aqui grafica los datos de nuestro dataset de arriba

#De aqui para abajo es puro ploting busines
gax.set_xlabel('longitude')
gax.set_ylabel('latitude')
gax.set_title('Target en Estados Unidos')

gax.spines['top'].set_visible(False)
gax.spines['right'].set_visible(False)

plt.show()
```

```python
geolocator = Nominatim(user_agent="kmeans-hw")

for index, row in geodataframe.iterrows():
    geodataframe.iloc[index, geodataframe.columns.get_loc('address')] = str(geolocator.reverse("{},{}".format(row.loc[1], row.loc[0])))

geodataframe.head()
```

```python
for index, row in geodataframe.iterrows():
    for index2, row2 in geodataframe.iloc[index + 1:].iterrows():
        print("Distancia entre ", row.iat[3], " y ", row2.iat[3])
        print(str(geodesic("{},{}".format(row.iat[1], row.iat[0]), "{},{}".format(row2.iat[1], row2.iat[0])).km), "kms")
```

```python
latlong['label'] = kmeans.labels_
stores = latlong["label"].value_counts()
print("Número de almacenes:", len(np.unique(latlong['label'])))
print("Número de tiendas por almacén:", stores)
```

###GeoPy###

GeoPy facilita a los desarrolladores de Python la localización de las coordenadas de direcciones, ciudades, países y puntos de referencia en todo el mundo.

¡Este módulo le permite recuperar puntos de latitud y longitud, pero también obtener códigos postales, distancias e incluso más! Entre las diversas bibliotecas para trabajar con datos geográficos, esta es definitivamente una de las más populares y famosas.


###Shapely###

Shapely es una biblioteca de geometría computacional basada en la biblioteca C/C++, GEOS (el motor de software detrás de PostGIS) y la biblioteca Java JTS. Permite el procesamiento geométrico de estilo PostGIS fuera de un sistema de administración de bases de datos relacionales y fuera de SQL. Shapely necesita combinarse con otras bibliotecas para leer y escribir archivos espaciales.


###PyProj###

Una biblioteca de funciones contenedoras alrededor de la biblioteca C PROJ (Biblioteca de Proyecciones Cartográficas y Transformaciones de Coordenadas). PyProj ofrece funcionalidad para convertir coordenadas de una referencia espacial a otra.


###GeoPandas###

GeoPandas amplía la muy útil biblioteca de Pandas, lo que permite el procesamiento de datos geométricos en Pandas DataFrames que, de lo contrario, requeriría una base de datos relacional espacial. GeoPandas depende de Fiona para las capacidades de lectura y escritura, y de Shapely para las funciones de geometría computacional.


###GeoMesa###

Otra biblioteca para Geoespacial escalable: GeoMesa ofrece capacidades para el procesamiento de datos vectoriales a gran escala a través de la computación distribuida. La transmisión de datos casi en tiempo real también es una opción para entidades vectoriales de puntos, líneas y polígonos.
