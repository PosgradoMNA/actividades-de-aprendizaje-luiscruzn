#**Análisis de varianza.** 

Supongamos que queremos analizar una variable categórica y ver la correlación entre diferentes categorías. Por ejemplo, considere el conjunto de datos del automóvil, la pregunta que podemos hacer es, ¿cómo las diferentes categorías de la característica Marca (como una variable categórica) tienen un impacto en el precio? El diagrama muestra el precio promedio de diferentes marcas de vehículos. Vemos una tendencia de precios crecientes a medida que avanzamos a lo largo del gráfico. Pero, ¿qué categoría en la función de marca tiene más y cuál tiene menos impacto en la predicción del precio del automóvil? Para analizar variables categóricas como la variable "hacer", podemos utilizar un método como el método ANOVA. 

ANOVA es una prueba estadística que significa "Análisis de varianza". ANOVA se puede utilizar para encontrar la correlación entre diferentes grupos de una variable categórica. De acuerdo con el conjunto de datos del automóvil, podemos usar ANOVA para ver si hay alguna diferencia en el precio medio de las diferentes marcas de automóviles, como Subaru y Honda. La prueba ANOVA devuelve dos valores: la puntuación de la prueba F y el valor p. La prueba F calcula la relación de variación entre la media de los grupos sobre la variación dentro de cada uno de los grupos de muestra. El valor p muestra si el resultado obtenido es estadísticamente significativo. La prueba F calcula la relación de variación entre las medias de los grupos sobre la variación dentro de cada una de las medias de los grupos de muestra. 

La prueba ANOVA se puede realizar en Python utilizando el método f_oneway como función integrada del paquete Scipy. Pasamos los datos de precios de los dos grupos de marcas de automóviles que queremos comparar y calcula los resultados de ANOVA. 

#**Correlación**

La correlación es una métrica estadística para medir en qué medida las diferentes variables son interdependientes. En otras palabras, cuando observamos dos variables a lo largo del tiempo, si una variable cambia, ¿cómo cambia este efecto en la otra variable? Por ejemplo, se sabe que fumar está relacionado con el cáncer de pulmón, ya que tiene una mayor probabilidad de contraer cáncer de pulmón si fuma. En otro ejemplo, existe una correlación entre las variables "Paraguas" y "Lluvia", donde más precipitación significa que más personas usan paraguas. Además, si no llueve, la gente no llevaría paraguas. Por lo tanto, podemos decir que los paraguas y la lluvia son interdependientes y por definición están correlacionados. Es importante saber que correlación no implica causalidad. De hecho, podemos decir que el paraguas y la lluvia están correlacionados, pero no tendríamos suficiente información para decir si el paraguas causó la lluvia o la lluvia causó el paraguas. 

En la ciencia de datos, generalmente nos ocupamos más de la correlación. Veamos la correlación entre el tamaño del motor y el precio. Esta vez, visualizaremos estas dos variables usando un diagrama de dispersión y una línea lineal adicional llamada "línea de regresión", que indica la relación entre las dos. El objetivo principal de este gráfico es ver si el "tamaño del motor" tiene algún impacto en el precio. En este ejemplo, puede ver que la línea recta a través de los puntos de datos es muy empinada, lo que muestra que existe una relación lineal positiva entre las dos variables. Con el aumento de los valores del tamaño del motor, los valores del precio también aumentan y la pendiente de la línea es positiva, por lo que existe una correlación positiva entre el "tamaño del motor" y el "precio". 

Podemos usar seaborn reregplot() para crear el diagrama de dispersión. Como otro ejemplo, ahora veamos la relación entre las millas de carretera por galón para ver su impacto en el "precio" del automóvil. Como podemos ver en este gráfico, cuando el valor de mpg en carretera sube, el valor del precio baja, por lo que existe una relación lineal negativa entre mpg en carretera y el precio. Aunque esta relación es negativa, la pendiente de la línea es pronunciada, lo que significa que las millas por galón en carretera siguen siendo un buen predictor del precio. Se dice que estas dos variables tienen una correlación negativa. Finalmente, tenemos un ejemplo de una correlación débil, por ejemplo, tanto las rpm máximas bajas como los valores altos de rpm máximas tienen precios altos y bajos, por lo tanto, no podemos usar RPM para predecir los valores.

#**Correlación estadística**

Una forma de medir la fuerza de la correlación entre variables numéricas continuas es usando un método llamado Correlación de Pearson. El método de correlación de Pearson le dará dos valores, el coeficiente de correlación y el valor p. Entonces, ¿cómo interpretamos estos valores? Para el coeficiente de correlación, un valor cercano a 1 implica una gran correlación positiva, mientras que un valor cercano a -1 implica una gran correlación negativa y un valor cercano a 0 implica que no hay correlación entre las variables. 

Luego, el valor p nos dirá cuán seguros estamos acerca de la correlación que calculamos. Para el valor p, un valor inferior a 0,001 nos da una gran certeza sobre el coeficiente de correlación que calculamos. Un valor entre 0,001 y 0,05 nos da una certeza moderada, un valor entre 0,05 y 0,1 nos dará una certeza débil y un valor de p mayor que 0,1 no nos dará ninguna certeza de correlación. Podemos decir que existe una fuerte correlación cuando el coeficiente de correlación es cercano a 1 o -1 y el valor de p es inferior a 0,001.
